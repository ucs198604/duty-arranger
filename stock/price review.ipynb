```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python

tikers = ["VT", "VTI",'RDS-B']
new_data = pd.DataFrame()
for i in tikers:
    new_data[i] = wb.DataReader(i, data_source="yahoo", start='2016-1-15')['Adj Close']  # RDSB buy at lowest price 
```

```python
new_data = new_data/new_data.iloc[0]
new_data.plot(figsize=(16,8))
```
